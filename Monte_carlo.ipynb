## Calcolare il $\pi$ con il metodo di Monte Carlo

Nasce nella *2 guerra mondiale* a *Los salomos* per il progetto *Manhattan* => **Bomba Atomica**


Generano numeri casuali per fare simulazioni o calcolare qualcosa

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_lanci = 10000
raggio = 1.

x = np.random.uniform(-raggio, raggio, n_lanci)
y = np.random.uniform(-raggio, raggio, n_lanci)

inside_circle = x**2 + y**2 < raggio**2
n_inside_circle = np.count_nonzero(inside_circle)

print(f"valore stimato: {4*n_inside_circle/n_lanci}")

pi = {4*n_inside_circle/n_lanci}
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(x,y,c=inside_circle, s= 1., cmap="bwr")

ax.set_title(r"il valore di $\pi$ stimato con %s : %s"%(n_lanci, pi))

plt.show()
```

```python
def stimaPi(n_lanci):
    raggio = 1.
    
    x = np.random.uniform(-raggio, raggio, n_lanci)
    y = np.random.uniform(-raggio, raggio, n_lanci)
    
    inside_circle = x**2 + y**2 < raggio**2
    n_inside_circle = np.count_nonzero(inside_circle)
    
    print(f"valore stimato: {4*n_inside_circle/n_lanci}")
    pi = 4*n_inside_circle/n_lanci
    return pi

n_lanci_array = np.array([10, 100, 1000, 10000, 100000, 1000000]) 
array = np.array([stimaPi(n) for n in n_lanci_array])
diff = np.abs(np.pi - array)
print(diff)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.plot(diff)
# ax.hlines(np.pi, 0, 10_000_000)
plt.show()
```
