## Calcolare il $\pi$ con il metodo di Monte Carlo

Nasce nella *2 guerra mondiale* a *Los salomos* per il progetto *Manhattan* => **Bomba Atomica**


Generano numeri casuali per fare simulazioni o calcolare qualcosa

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_lanci = 10000
raggio = 1.

x = np.random.uniform(-raggio, raggio, n_lanci)
y = np.random.uniform(-raggio, raggio, n_lanci)

inside_circle = x**2 + y**2 < raggio**2
n_inside_circle = np.count_nonzero(inside_circle)

print(f"valore stimato: {4*n_inside_circle/n_lanci}")

pi = {4*n_inside_circle/n_lanci}
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(x,y,c=inside_circle, s= 1., cmap="bwr")

ax.set_title(r"il valore di $\pi$ stimato con %s : %s"%(n_lanci, pi))

plt.show()
```

```python
def stimaPi(n_lanci):
    raggio = 1.
    
    x = np.random.uniform(-raggio, raggio, n_lanci)
    y = np.random.uniform(-raggio, raggio, n_lanci)
    
    inside_circle = x**2 + y**2 < raggio**2
    n_inside_circle = np.count_nonzero(inside_circle)
    
    print(f"valore stimato: {4*n_inside_circle/n_lanci}")
    pi = 4*n_inside_circle/n_lanci
    return pi

n_lanci_array = np.array([10, 100, 1000, 10000, 100000, 1000000]) 
array = np.array([stimaPi(n) for n in n_lanci_array])
diff = np.abs(np.pi - array)
print(diff)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.plot(diff)
# ax.hlines(np.pi, 0, 10_000_000)
plt.show()
```

```python
def f(x):
    return x

N_LANCI = 1000
a = 1.
b = 10.
x = np.linspace(a, b, 1000)
y = f(x)

x_punti = np.random.uniform(a, b, N_LANCI)
y_punti = np.random.uniform(-1, np.max(y), N_LANCI)

area_rect = (b - a) * np.max(y)
inside_area = f(x_punti) > y_punti
n_inside = np.count_nonzero(inside_area)

area_f = n_inside / N_LANCI * area_rect

fig, ax = plt.subplots(figsize = (6,3))
ax.plot(x,y, label="$f(x)", color= np.array([1,0,1]))
ax.grid()
ax.scatter(x_punti,y_punti, c = inside_area, s = 2, label="punto monte carlo")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend()
ax.set_title(f"l'area stimata  {area_f:.2f}")
fig.set_facecolor("pink")
plt.show()
```

## Grafici a dispersione
* con `plot()` non possiamo personalizzare i singoli punti
* con `scatter()` possiamo personalizzare i singoli punti

```python
N_PUNTI = 20
x = np.random.uniform(0., 1.,N_PUNTI)
y = np.random.uniform(0., 1.,N_PUNTI)

color = np.random.uniform(0.,1., N_PUNTI)
size = np.random.uniform(10.,100., N_PUNTI)
fg, ax = plt.subplots(figsize = (4,4))
ax.scatter(x,y, c = color, s = size,  cmap="inferno", edgecolors="k")
plt.show()
```
