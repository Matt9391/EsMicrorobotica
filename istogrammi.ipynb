```python
import numpy as np
import matplotlib.pyplot as plt
```

# Grafici: ISTROGRAMMI
gli istogrammi non sono grafici a barre

```python
NPEOPLE = 1000
heights = np.round(np.random.normal(170, 12,NPEOPLE),1) #in CM
```

```python
fig, ax = plt.subplots(figsize = (4,3))
ax.hist(heights, bins = 20) #bins is the number of interval of x ace
ax.set_xlabel("Height [cm]")
ax.set_ylabel("Num of people")
plt.show()
```

```python
N_LANCI = 10000
N_DADI = 10000
punteggio = np.random.randint(1,7, N_LANCI)
for i in range (1, N_DADI):
    dice = np.random.randint(1,7, N_LANCI)
    punteggio += dice

fig, ax = plt.subplots(figsize = (4,3))

ax.hist(punteggio, bins = 11)
ax.set_xlabel("punteggio")
ax.set_ylabel("numero di volte in cui Ã¨ uscito un punteggio")

plt.show()
                    
```

```python
N_LANCI = 10000
N_DADI = 10000

dice = np.random.randint(1,7, (N_DADI, N_LANCI))
punteggio = np.sum(dice, axis=0)

fig, ax = plt.subplots(figsize = (4,3))

ax.hist(punteggio, bins = 11)
ax.set_xlabel("punteggio")
ax.set_ylabel("numero di volte in cui Ã¨ uscito un punteggio")

plt.show()
                    
```

```python
N_LANCI = 10000
N_DADI = 10000

def lancio_dadi(n_lanci, n_dadi):
    dice = np.random.randint(1,7, (n_dadi, n_lanci))
    return np.sum(dice, axis=0)

dadi = [2,5,10]
n_intervalli = [11, 15, 20]
punteggio = [lancio_dadi(N_LANCI, n) for n in dadi]
print(punteggio)
fig, ax = plt.subplots(figsize = (4,3))

ax.set_xlabel("punteggio")
ax.set_ylabel("numero di volte in cui Ã¨ uscito un punteggio")

for p, bins in zip(punteggio, n_intervalli):
    ax.hist(p,alpha = 1, bins = bins)

plt.show()

```
