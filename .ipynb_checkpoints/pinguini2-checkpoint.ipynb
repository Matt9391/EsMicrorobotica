```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("penguins_size.csv")
cols = ["species","island","culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g", "sex"]
```

```python
def scatter_ax(ax,p1,p2):
        species = {'Adelie' : 'red', 'Chinstrap' : 'green', 'Gentoo' : 'blue'}
        for s in species:
            specie_data = data[data["species"] == s]
            ax.scatter(specie_data[p1],specie_data[p2], color = species[s], alpha = 0.4, s = 45, label = s)
            ax.set_xlabel(p1)
            ax.set_ylabel(p2)
            # ax.legend()
```

```python
fig, axs= plt.subplots(nrows = 6, ncols = 6, figsize = (15,20))

for i in range(0,6):
    for j in range(0,6):
        scatter_ax(axs[i][j], cols[i], cols[j])

plt.show()
```
