```python
#Es 2

n = 100

array = np.ones(n, dtype=bool)
array[:2] = False #imposto a false i primi 2 numeri perché non primi
for i in range(2, int(n**0.5) + 1):
    if array[i]:
            # parti dal prossimo multiplo e scorri fino alla fine
            # e a ogni passo di "i" cioè il numero primo attuale, quindi un multiplo lo togli
            array[i*i : n : i] = False
array = np.nonzero(array)[0]
print(array)

```

```python
def f1(x):
    return np.sin(x)**2
def f2(x):
    return np.e**(-x**2)

N_LANCI = 1000
a = -5.
b = 5.
x = np.linspace(a, b, 1000)
y1 = f1(x)
y2 = f2(x)

fig, ax = plt.subplots(figsize = (6,3))
ax.plot(x,y1, label="$\sin(x^2)$", color= "#e55812")
ax.plot(x,y2, label="$e^{-x^2}$", color= "#97c726", linestyle = "--")
ax.grid()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend()
ax.set_title(f"Grafico delle funzioni f(x) e g(x)")
ax.set_facecolor("#efe7da")
plt.show()
```
