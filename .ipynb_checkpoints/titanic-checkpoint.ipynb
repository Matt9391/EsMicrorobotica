```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("titanic.csv")
data.head(5)
data["Pclass"].value_counts()
data[["Pclass", "Age"]].groupby("Pclass").count()

data[["Pclass", "Fare"]].groupby("Pclass").mean()

data[["Sex", "Survived", "Age"]].groupby(["Sex", "Survived"]).count()

data[data["Age"] < 14]["Survived"].value_counts()

data["hasRelative"] = [0 if d == 0 else 1 for d in data["Parents/Children Aboard"]]

data[["hasRelative", "Survived"]].value_counts()
```

```python
def scatterAx(ax, df, cat, nameCat, p1, p2):
    for c in cat:
        data = df[df[nameCat] == c]
        ax.scatter(data[p1], data[p2], color = cat[c], alpha = 0.4, s = 45, label = c)

def histAx(ax, df, cat, nameCat, p1):
    for c in cat:
        data = df[df[nameCat] == c]
        ax.hist(data[p1], bins = 15, alpha = 0.5)
```

```python
cols = data.columns.unique()
print(cols)

fig, axs = plt.subplots(nrows = 6, ncols = 6, figsize = (32,32))

for i in range(0,6):
    for j in range(0,6):
        if(i != j):
            scatterAx(axs[i][j], data, cols, "Survived", cols[i],cols[j])
```
