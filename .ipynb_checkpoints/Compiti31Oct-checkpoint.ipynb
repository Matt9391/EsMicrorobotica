## Es 9

```python
p1 = 0
p2 = 0
for i in range(6):
    dice1 = np.random.randint(1,7,1000)
    dice2 = np.random.randint(1,7,1000)
    draw = dice1 == dice2
    p1 += np.count_nonzero(draw) / 2.
    p2 += np.count_nonzero(draw) / 2.
    p1wins = dice1[dice1 != dice2] > dice2[dice1 != dice2]
    p1 += np.count_nonzero(p1wins)
    p2 += len(dice2[dice1 != dice2]) - np.count_nonzero(p1wins)
print(p1,p2)
```

```python
NPARTITE = 6
NLANCI = 10000
dice1 = np.random.randint(1,7,(NPARTITE, NLANCI))
dice2 = np.random.randint(1,7,(NPARTITE, NLANCI))
draws = dice1 == dice2

wins1 = dice1 > dice2

points1 = wins1*1. + draws*0.5
gamePoints1 = np.sum(points1, axis=1)
gamePoints2 = NLANCI - gamePoints1
print(gamePoints1)
print(gamePoints2)
```

```python
NPARTITE = 6
NLANCI = 10000
nums = np.arange(1,7)
probs = [1/8.,1/8.,1/8.,1/8.,2/8.,1/4.]
dice1 = np.random.choice(nums,(NPARTITE, NLANCI), p=probs)
dice2 = np.random.choice(nums,(NPARTITE, NLANCI))

draws = dice1 == dice2

wins1 = dice1 > dice2

points1 = wins1*1. + draws*0.5
gamePoints1 = np.sum(points1, axis=1)
gamePoints2 = NLANCI - gamePoints1
print(gamePoints1)
print(gamePoints2)
```

## Es 10

```python
N = 1000000
mean = 19
varianza = 5
vStd = np.sqrt(varianza)
err = 20

array = np.random.normal(mean, vStd, N)
error_indices = np.random.choice(np.arange(0,N), err, replace=False)
array[error_indices] = 9999.9 #fancy indicizzazione

cleanArray = array[array != 9999.9]

meanArr = np.mean(cleanArray)
maxArr = np.max(cleanArray)
minArr = np.min(cleanArray)

print(meanArr,maxArr,minArr)



array_1 = np.roll(array,1) == 9999.9
for i in range(2,6):
    array_1 = array_1 & (np.roll(array,i) == 9999.9)
rotta = np.count_nonzero(array_1)
print(rotta)


fix, ax = plt.subplots(figsize = (5,2))
#Rappresentiamo la distribuzione dei valori in array pulisto
ax.hist(cleanArray, bins=50)
ax.set_xlabel("valore pm10")
ax.set_ylabel("numero giorni")
plt.show()
```

```python
SIZE = (200,200, 3)
img = np.zeros(SIZE)
x = np.arange(0,SIZE[0])
y = np.arange(0,SIZE[0])
x_matrix, y_matrix = np.meshgrid(x,y)
#print(x_matrix)
#print(y_matrix)
cerchio = ((x_matrix-SIZE[0]//2)**2 + (y_matrix-SIZE[1]//2)**2) <= 50**2 
img[cerchio] = [1,0,0]
plt.imshow(img)
```

```python
N = 1000000
array_random = np.random.uniform(0, 1.,N)
fix, ax = plt.subplots(figsize = (5,2))
#Rappresentiamo la distribuzione dei valori in array pulisto
ax.hist(array_random, bins=50)
ax.set_xlabel("numero random")
ax.set_ylabel("frequenza")
plt.show()
```
