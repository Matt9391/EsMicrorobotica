```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("./penguins_size.csv")
data # oggetto di tipo dataframe
```

```python
data.dropna(inplace = True) # NON andrebbe fatta così
data
```

```python
dataArray = data[["culmen_length_mm","culmen_depth_mm","flipper_length_mm","body_mass_g"]].to_numpy()
species = data["species"].to_numpy()
```

```python
dataArray.shape
dataArray.dtype
```

```python
# print(species)
species2num = {"Adelie" : 0, "Chinstrap" : 1, "Gentoo" : 2}
speciesNum = np.zeros(len(species), dtype = np.int8)
for s in species2num:
    speciesNum[species == s] = species2num[s]
speciesNum
```

```python
x = dataArray[:, 2]
y = dataArray[:, 3]
z = dataArray[:, 1]

fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (8,6))

axs[0, 0].scatter(x,y, edgecolor = "k", alpha = 0.7, c = speciesNum)
axs[0, 0].set_xlabel("flipper length [mm]")
axs[0, 0].set_ylabel("body mass [g]")

axs[0, 1].scatter(x,z, edgecolor = "k", alpha = 0.7, c = speciesNum)
axs[0, 1].set_xlabel("flipper length [mm]")
axs[0, 1].set_ylabel("culment lenght [mm]")

axs[1, 0].hist(y,edgecolor = "k", alpha = 0.7)
axs[1, 0].set_xlabel("body mass [g]")
axs[1, 0].set_ylabel("n")

axs[1, 1].hist(x,edgecolor = "k", alpha = 0.7)
axs[1, 1].set_xlabel("culment lenght [mm]")
axs[1, 1].set_ylabel("n")
plt.show()
```
