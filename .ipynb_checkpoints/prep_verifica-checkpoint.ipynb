```python
import numpy as np

NALUNNI = 27
NVOTI = 10
MINVOTO = 2
MAXVOTO = 10

matrix = np.random.randint(MINVOTO, MAXVOTO +1, (NALUNNI, NVOTI))
meanAlumns = np.mean(matrix, axis=1)
print(np.count_nonzero(meanAlumns > 6))
```

```python
indici  = np.arange(1, NALUNNI + 1)
print(indici)

indici[meanAlumns > 6]
```

```python
import matplotlib.pyplot as plt

ROWS = 10
COLS = 2 * ROWS

matrix = np.zeros((ROWS, COLS))
matrix[::2, COLS//2::2] = 10
matrix[1::2, COLS//2 + 1::2] = 10

matrix[:, :COLS//2] = np.random.randint(1, 10, (ROWS, COLS//2))

plt.imshow(matrix, cmap="inferno")
```

```python
import matplotlib.pyplot as plt

N = 5
SIZE = (10,20)

matrix = np.zeros(SIZE, dtype = np.int8)

x = np.random.randint(0, SIZE[0], N)
y = np.random.randint(0, SIZE[1], N)

matrix[x,y] = 1 # indicizzazione fancy
# print(matrix)

plt.imshow(matrix)
```

```python
N = 399
SIZE = (20,20)

array = np.zeros(SIZE[0] * SIZE[1], dtype = np.int8)
indices = np.arange(0, SIZE[0] * SIZE[1])
choosed_indices = np.random.choice(indices, N, replace= False)
array[choosed_indices] = 1

matrix = array.reshape(SIZE)

plt.imshow(matrix, cmap="Greys")
```
