```python
#selezionare i record con emotion = 0
angry = data[data["emotion"] == 0]
angry_face_array = angry["pixels"].to_numpy()
angry_face_array.shape
```

```python
test_face = angry_face_array[0]
test_face_array = np.array([int(p) for p in test_face.split(" ")]).astype(np.uint8)
```

```python
emotion_dict = {
    0: "Angry",
    1: "Disgust",
    2: "Fear",
    3: "Happy",
    4: "Sad",
    5: "Surprise",
    6: "Neutral"
}

fig, axs = plt.subplots(nrows = 7, ncols = 7,  figsize=(20,3))

for i in range(0,7):
    faces = data[data["emotion"] == i]
    faces_array = faces["pixels"].to_numpy()
    for j in range(0,7):
        face = faces_array[j]
        face_array = np.array([int(p) for p in face.split(" ")]).astype(np.uint8)
    
        axs[i][j].set_title(emotion_dict[i])
        
        axs[i].imshow(face_array.reshape(48,48), cmap="binary_r") #immagine in scala di grigia 48x48
        axs[i].axis('off')

plt.show()
```
