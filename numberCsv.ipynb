```python
data = pd.read_csv("./mnist_784_csv.csv")
pixel_columns = list(data.columns[:-1])
```

```python
fig, axs = plt.subplots(nrows = 10, ncols = 10, figsize=(10,10))

for i in range(0, 10):
    data_numbers = data[data['class'] == i] #Dataframe
    images_number = data_numbers[pixel_columns].to_numpy().astype(np.uint8) #array np
    print(images_number.shape)
    indexes = np.random.choice(np.arange(0,len(images_number)),10, replace = False)
    axs[i][5].set_title(i)
    
    for j in range(0, 10):
        axs[i][j].imshow(images_number[indexes[j]].reshape(28,28), cmap = "binary")
        axs[i][j].axis('off')    

plt.show()
```
