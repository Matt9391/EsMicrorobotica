# Simuliamo il prezzo giornaliero di una azione

```python
import numpy as np
import matplotlib.pyplot as plt
N_GIORNI = 100
variazione_prezzo = np.random.normal(-0.001,5, N_GIORNI)
prezzo_iniziale = 50
# print(variazione_prezzo)
prezzo_giornaliero = prezzo_iniziale + np.cumsum(variazione_prezzo)
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(prezzo_giornaliero)

# ricerca massimi e minimi locali
arrPrev = np.roll(prezzo_giornaliero, -1)
arrNext = np.roll(prezzo_giornaliero, 1)
arrLocal = prezzo_giornaliero
# arrLocal = cleanArray[1:-1]

localMaxBoolIndices = (prezzo_giornaliero > arrPrev) & (prezzo_giornaliero > arrNext)
print(localMaxBoolIndices.shape)
arrLocal[localMaxBoolIndices]

print(arrLocal[localMaxBoolIndices])
plt.show()
```
